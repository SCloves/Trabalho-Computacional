# 1. Intervalo de confiança para uma amostra. #

#### Gere n = 1000 amostras de tamanho m = 40 de uma população Normal(μ, σ 2 ), para μ e σ² fixos.

```r
# como não foi especificado  μ e σ²
# vou usar a normal padronizada (0, 1)

numero_de_amostras = 1000
tamanho_de_cada_amostra = 40

lista_de_amostras <- list(mode="vector",length=numero_de_amostras)

for (i in 1 : numero_de_amostras) 
{
  lista_de_amostras[[i]] <- rnorm(n = tamanho_de_cada_amostra, mean = 0, sd = 1)
}
```

```r
# valores da primeira amostra
lista_de_amostras[[1]]
```

#### Obtenha as estimativas X̄ e S 2 para μ e σ² , respectivamente, utilizando cada uma das amostras.

```r
# obtendo a média de cada amostra

estimativas_das_medias <- 0

for (i in 1 : numero_de_amostras) 
{
   estimativas_das_medias[i] <- mean(lista_de_amostras[[i]])
}

estimativas_das_variacias <- 0 

for (i in 1 : numero_de_amostras) 
{
   estimativas_das_variacias[i] <- var(lista_de_amostras[[i]])
}
```

#### A partir das estimativas obtidas no item anterior, construa um histograma de X̄ e outro de S² .

```r
hist(estimativas_das_medias, main='Histograma das Estimativas das Medias',
    xlab='Estimativas das Medias', ylab='Frequência', col='blue')
```

```r
hist(estimativas_das_variacias, main='Histograma das Estimativas das Variâncias',
    xlab='Estimativas das Variâncias', ylab='Frequência', col='green')
```
